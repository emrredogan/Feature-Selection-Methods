### Import Libraries

```python
import sys 
sys.path.append('/home/emred/feature_selection_project/src')
from feature_selection import FeatureSelection
import pandas as pd
from IPython.display import display
```

### Load Data

```python
df = pd.read_csv('/home/emred/feature_selection_project/input/lung_cancer.csv', index_col='index')
df.head(5)
```

### EDA & Feature Engineering

```python
df = FeatureSelection.df_rename(df=df)
display(df.head(5))
```

```python
df_desriptive_stats = FeatureSelection.descriptive_stats(df=df)
display(df_desriptive_stats)
```

```python
display(FeatureSelection.check_missing(df=df))
```

### Correlation Map & Multicollinerity Elimination

```python
display(FeatureSelection.df_corr_heatmap(df=df))
```

```python
df_after_multicollinearity = FeatureSelection.df_multicollinearity_elimination(df=df)
print(f"df columns number: {len(df.columns)}")
print(f"df columns number after multicollinearity elimination: {len(df_after_multicollinearity.columns)}")
```

### Train-test Split 

```python
X_train, X_test, y_train, y_test = FeatureSelection.train_test_split(df=df_after_multicollinearity)
print("\n X_train shape: ", X_train.shape, "\n X_test shape: ", X_test.shape, "\n y_train shape: ", y_train.shape, "\n y_test shape: " , y_test.shape)
```

### Create FeatureSelection Class Object 

```python
log_class = FeatureSelection(df=df_after_multicollinearity, X_train=X_train, X_test= X_test, y_train=y_train, y_test=y_test)
```

### SequentialFeatureSelector (forward) Method

```python
sfs_best_feature_names, sfs_best_feature_idx, df_step_forward_table =log_class.step_forward_feature_selection()
```

```python
print("sfs_best_feature_names: ", sfs_best_feature_names, "\n sfs_best_feauture_idx: ", sfs_best_feature_idx)
```

```python
display(df_step_forward_table)
```

### ExhaustiveFeatureSelector Method

```python
efs_best_feature_names, efs_best_feature_idx, df_exhaustive_table = log_class.exhaustive_selection(X_train=X_train)
```

```python
print("efs_best_feature_names: ", efs_best_feature_names, "\n efs_best_feauture_idx: ", efs_best_feature_idx)
```

```python
display(df_exhaustive_table)
```

### Logistic Regression Models with Selected Features

```python
feature_names = list(X_train.columns)
selected_features = efs_best_feature_idx
selected_feature_names = [feature_names[i] for i in selected_features]
print(selected_feature_names)
```

```python
sequential_logistic = log_class.logistic_reg_with_best_feat(X_train=X_train.loc[:,sfs_best_feature_names])
exhaustive_logistic = log_class.logistic_reg_with_best_feat(X_train=X_train.loc[:,selected_feature_names])
```

```python
sequential_performance = log_class.performance_logistic(clf=sequential_logistic, X_test=X_test.loc[:,sfs_best_feature_names])
exhaustive_performance = log_class.performance_logistic(clf=exhaustive_logistic, X_test=X_test.loc[:,selected_feature_names])
```

```python
print(f"Gini value of logistic regression with Sequential forward method is {sequential_performance}")
print(f"Gini value of logistic regression with exhaustive method is {exhaustive_performance}")
```
